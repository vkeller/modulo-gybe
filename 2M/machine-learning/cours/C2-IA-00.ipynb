## Conclusion

On peut donc établir les différentes étapes d'une intelligence artificielle de classification :

1. **Labelisation** des images
1. **Entraînement**. C'est la phase d'apprentissage du réseeau de neurones. C'est à cette étape que l'on utilise un **Ensemble de données d'entraînement** (en anglais *Training dataset*)
1. **Validation**, C'est la phase de fine tuning du modèle. On utilise un **Ensemble de données de validation** (en anglais *Validatation dataset*)
1. **Test**. On teste le modèle sur un ensemble inconnu et non labellisé. On utilise à ce stade un **Ensemble de données de test** (en anglais *Test dataset*)
