# TP2-ESDN-2026 : Enjeux sociaux du numérique

## Objectifs

- Présenter les enjeux sociaux ainsi que les objectifs de l'évaluation
- Comprendre la méthode
- Créer les groupes de 2 ou 3 élèves
- Choisir les dates des exposés

## Méthode

Elle se déroulera en 7 étapes:

1. Choisir un sujet pour l'exposé ayant trait aux enjeux sociaux du numérique
2. Recherche d'informations
3. Ecrire son exposé
4. Utilisation de l'IA générative pour évaluer son exposé
5. Préparer 8 questions
6. Présenter son exposé et tester ses camarades
7. Prendre des notes

Regroupez toutes vos informations et sources dans un document word. Ce document word (exporté en PDF) doit **contenir au minimum 3 sources fiables**

### 3. Ecrire son exposé

A partir des sources que vous aurez lues, écrivez une première ébauche de votre exposé.

**Il existe des outils d'IA pour cette étape**, [Google Notebook LM](https://notebooklm.google/) en est un ! NotebookLM est une IA qui permet des productions sur la base de sa propre liste de sources fiables. 

**LISEZ** votre exposé et vérifiez qu'il ne contient pas d'erreur. Produisez-en un PDF

### 4. Auto-évaluation (IAg)

Demandez à une IAg (ChatGPT, QWen, Perplexity, Gemini, etc..) d'évaluer votre production. Vous pouvez utiliser un prompt tel que :

> Tu es un professeur dans un gymnase. Tu dois superviser un travail fait par un élève sur le sujet "Enjeux sociaux du numérique". Tu vas évaluer son travail (un fichier pdf que je vais t'envoyer) selon 3 critères :
> 1. Quantité d'information : y a-t-il suffisamment de matières pour faire un exposé de 10 minutes? En particulier, y a-t-il des exemples concrets ?
> 2. Qualité du contenu : exactitude, pertinence
> 3. Organisation et clarté
>
> Pour chacun des critères, justifie brièvement tes points.

### 5. Préparer 8 questions

Il vous faudra trouver 8 questions assez difficiles à poser à la classe une fois l'exposé terminé. A nouveau ChatGPT peut vous aider : 

> Voici un pdf de mon exposé. Trouve 8 questions assez difficiles à partir de ce document. Propose 4 réponses plausibles pour chaque question.

Consigne : 

1. Préparer un Powerpoint de 16 slides (question 1, réponse 1, question 2, réponse 2, etc..)
2. Laisser le temps aux camarades de répondre

**L'enseignant posera lui aussi des questions**

### 6. Présenter son exposé

Vous avez le droit à votre feuille d'exposé, mais il ne doit pas être lu.

Timing :

- 10 min de présentation
- 4 min pour les questions (30 secondes par question)

### 7. Prendre des notes

C'est le point **crucial**. Les auditeurs (les camarades de la classe) devront prendre des notes dans un cahier spécifique acquis pour l'occasion. 

**Ce cahier sera le seul matériel autorisé pour l'évaluation !!!**

La prise de notes manuscrite présente plusieurs avantages, notamment celui de mémoriser plus facilement. 

## Traces à envoyer à l'enseignant (20h LE SOIR PRÉCÉDENT L'EXPOSÉ)

1. Le PDF de l'exposé
2. Le PDF avec au minimum 3 sources
3. Le Powerpoint avec les 8 questions et les 8 réponses qui se suivent
4. L'auto-évaluation de ChatGPT (prompt + réponse de l'IAg)

## EVALUATION

L'évaluation se fera en classe sous la forme d'un QCM. Le QCM sera formé d'un choix de 3 questions par exposé. S'il y a eu 8 groupes, alors il y aura 24 questions

Seul matériel autorisé : **le cahier de prises de notes**
